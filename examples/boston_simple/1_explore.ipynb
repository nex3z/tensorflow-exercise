```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
df_single = pd.read_csv('./data/data_single.csv', names=['area', 'price'])
df_single.head()
```

```python
sb.set(context='notebook', style='whitegrid')
sb.lmplot('area', 'price', df_single, height=6, fit_reg=True)
```

```python
df_multi = pd.read_csv('./data/data_multi.csv', names=['area', 'bedrooms', 'price'])
df_multi.head()
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.gca(projection='3d')
ax.scatter(df_multi.area, df_multi.bedrooms, df_multi.price, c=df_multi.price, cmap='inferno')
ax.set_xlabel('area')
ax.set_ylabel('bedrooms')
ax.set_zlabel('price')
```

```python
df_norm = df_multi.apply(lambda column: (column - column.mean()) / column.std())
df_norm.head()
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.gca(projection='3d')
ax.scatter(df_norm.area, df_norm.bedrooms, df_norm.price, c=df_norm.price, cmap='inferno')
ax.set_xlabel('area')
ax.set_ylabel('bedrooms')
ax.set_zlabel('price')
```
