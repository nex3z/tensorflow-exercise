```
import tensorflow as tf
import tensorflow_datasets as tfds
import tensorflow_hub as hub
from tensorflow import keras
```

```
BATCH_SIZE = 32
```

```
dataset, info = tfds.load('imdb_reviews', as_supervised=True, with_info=True)
train_data, test_data = dataset['train'], dataset['test']
train_size = info.splits['train'].num_examples
test_size = info.splits['test'].num_examples
print(f"train_size = {train_size}, test_size = {test_size}")
```

```
train_set = train_data.repeat().batch(BATCH_SIZE).prefetch(1)
```

```
model = keras.Sequential([
    hub.KerasLayer('https://tfhub.dev/google/tf2-preview/nnlm-en-dim50/1', dtype=tf.string, input_shape=[], output_shape=[50]),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```
history = model.fit(train_set, steps_per_epoch=train_size//BATCH_SIZE, epochs=5)
```

```
for x_batch, y_batch in test_data.batch(1).take(1):
    print(f"x_batch = \n{x_batch}")
    print(f"y_batch = \n{y_batch}")
```

```
model.predict(x_batch)
```
